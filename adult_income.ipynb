```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('datasets/adult.csv')
data
```

```python
data = data.replace('?', np.nan)
```

```python
data.isnull().sum()
```

```python
data = data.dropna(axis= 0)
```

```python
data
```

```python
from sklearn.preprocessing import OneHotEncoder
categorical_col = ['workclass', 'education', 'occupation']
ohe = OneHotEncoder(drop= 'first', sparse_output= False)
encoded = ohe.fit_transform(data[categorical_col])
encoded_df = pd.DataFrame(encoded, columns= ohe.get_feature_names_out(categorical_col), index= data.index)
encoded_df
```

```python
merged_data = pd.concat([data, encoded_df], axis= 1)
merged_data
```

```python
sex = ['sex']
encoding = ohe.fit_transform(data[sex])
sex_encoded = pd.DataFrame(encoding, columns= ohe.get_feature_names_out(sex),index= data.index)
merged_data = pd.concat([merged_data,sex_encoded], axis= 1)
merged_data
```

```python
new = merged_data.drop(columns= ['workclass','education','marital.status','relationship','race'])
new
```

```python
new = new.drop(columns=['native.country'])
new
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
new['income'] = le.fit_transform(new['income'])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report
```

```python
X = new.drop(columns=['income'])
y = new.income
X_train, X_test, y_train, y_test = train_test_split(X,y,stratify=y,test_size=0.3,random_state=42)
```

```python
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
accuracy_score(y_test, tree.predict(X_test))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
tree_param = {
    'max_depth':[None, 10,20],
    'min_samples_split':[2,3,5,10],
    'min_samples_leaf':[2,5,10],
    'max_leaf_nodes':[None,10,20,30]
}
tree_grid = GridSearchCV(
    estimator=tree,
    param_grid=tree_param,
    cv= 5,
    n_jobs= -1,
    verbose= 1
)
tree_grid.fit(X_train, y_train)
```

```python
tree_param = {
    'max_depth':[3, 5, 7, 10, 15, 20, None],
    'min_samples_split':[2,3,5,10],
    'min_samples_leaf':[2,5,10],
    'max_leaf_nodes':[None,10,20,30,50,100]
}
tree_grid = GridSearchCV(
    estimator=tree,
    param_grid=tree_param,
    cv= 5,
    n_jobs= -1,
    verbose= 1
)
tree_grid.fit(X_train, y_train)
```

```python
best_model_tree = tree_grid.best_estimator_
y_pred_tree = best_model_tree.predict(X_test)
```

```python
print(f"Accuracy Score: {accuracy_score(y_test, y_pred_tree)}")
print(f"Precision: {precision_score(y_test, y_pred_tree)}")
print(f"Recall: {recall_score(y_test, y_pred_tree)}")
```

```python
print(classification_report(y_test, y_pred_tree))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred_tree))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
accuracy_score(y_test, rf.predict(X_test))
```

```python
rf_param = {
    'n_estimators':[100,200,300],
    'max_depth':[None, 10,20],
    'min_samples_split' : [2,3,5,10],
    'min_samples_leaf':[1,2,3,4],
    'max_features':['sqrt', 'log2']
}

rfcv = GridSearchCV(
    estimator= rf,
    param_grid=rf_param,
    cv= 3,
    n_jobs= -1,
    verbose= 1
)
rfcv.fit(X_train, y_train)
best_model_rf = rfcv.best_estimator_
```

```python
y_pred_rf = best_model_rf.predict(X_test)
accuracy_score(y_test, y_pred_rf)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
accuracy_score(y_test, lr.predict(X_test))
```

```python
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBClassifier

xg_param = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 4, 5, 6, 7],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'gamma': [0, 1, 5],
    'reg_alpha': [0, 0.1, 1],
    'reg_lambda': [1, 1.5, 2]
}

random_search = RandomizedSearchCV(
    estimator=XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    param_distributions=xg_param,
    n_iter=50,
    cv=3,
    scoring='accuracy',
    verbose=1,
    n_jobs=-1
)

random_search.fit(X_train, y_train)

print("Best Parameters:", random_search.best_params_)
xgm = random_search.best_estimator_
accuracy_score(y_test, xgm.predict(X_test))
```

```python
print("Report for XGBoost: \n")
print(classification_report(y_test, xgm.predict(X_test)))
```
